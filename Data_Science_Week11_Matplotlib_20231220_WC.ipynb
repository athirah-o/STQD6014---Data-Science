### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots



```python
# create one or more subplots
# 2 x 2; 4 plots in total

rng = np.random.RandomState(12345)
fig = plt.figure(dpi=300)
ax1 = fig.add_subplot(2,2,1) # -> (2,2) subplot of 2x2; "1" -> first plot, which is top left -> histogram
ax2 = fig.add_subplot(2,2,2) # -> (2,2) subplot of 2x2; "2" -> first plot, which is top right -> scatter plot
ax3= fig.add_subplot(2,2,3) # -> (2,2) subplot of 2x2; "3" -> first plot, which is bottom left -> line plot

#dont need plt function  because we have already define in fig variable
ax1.hist(rng.standard_normal(100), bins=20, color='maroon', alpha=0.8)
ax2.scatter(np.arange(30), np.arange(30) + 3*rng.standard_normal(30))
ax3.plot(rng.standard_normal(50).cumsum(), color='orange', linestyle='dashed');

#how to save figure
plt.savefig("/content/drive/MyDrive/STQD6014 - Data Science/Images/Image_1") #to save the figure in file
```

### **Adjusting the spacing around subplots**

```python
# An example that shrink the spacing of the subplots to zero
# change the value of wspace and hspace
# sharex=True parameter means that the subplots will share the same x-axis
# The "fig" variable is a reference to the entire figure
# The "axes" is a 2D array of subplot references
fig, axes = plt.subplots(2,2, sharex = True, sharey = False)

# The outer loop (for i in range(2)) iterates over rows
# The inner loop (for j in range(2)) iterates over columns
# Top left -> Top right -> bottom left -> bottom right
for i in range(2):
  for j in range(2):
    axes[i,j].hist(rng.standard_normal(500), bins=50, color="navy", alpha=0.5)

#Adjust space around the subplot
#wspace = width of the subplot
#hspace = height of the subplot
fig.subplots_adjust(wspace=0.5, hspace=0.5)
```

# **Adding shapes to a plot**
<font color = red> ***IMPORTANT FOR FINAL EXAM***

```python
# Adding shapes to a plot
fig, ax = plt.subplots()

#create rectange shape
#0.4 -> width of the x-axis
#0.2 -> 0.75 + 0.2 = 0.95 -> height of the rectangle
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.2, color="green", alpha=0.5)

#create Circle shape
#(0.7,0.2) -> center of the circle
#0.15 -> radius of the circle
circ = plt.Circle((0.7,0.2), 0.15, color="#ff4500", alpha=0.5)

#create Triangle shape
#[[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]] -> the 3 points of the triangle
trig = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color="#800080", alpha=0.5)

# all shape together in a figure
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(trig);
```

```python
# Adding shapes to a plot : Attention
import matplotlib.patches as mpatches
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
tri = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color="green", alpha=0.5)

# crate a Square shape
sqr = plt.Rectangle((0.3, 0.1), 0.2, 0.2, color="black", alpha=0.9)

# add ellipse
# (0.5, 0.5) -> This tuple specifies the center of the ellipse
# 0.2 -> width of the ellipse
# 0.1 -> height of the ellipse
ellipse = mpatches.Ellipse((0.5, 0.5), 0.2, 0.1)

# add hexagon
# 6 -> specify number of sides in the regular polygon
# 0.1 -> distance from the center of the circumcircle (the point at the center of the hexagon) to any of its vertices
hex = mpatches.RegularPolygon((0.7, 0.8), 6, 0.1)

# all shape together in a figure
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(tri)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hex);
```

# **Dealing with Data Frame: <color font = red> Important for final exam**

```python
# A little project: dataset about restaurant tipping
# Download the tips data herehttps://bit.ly/3jCROII:

tips = pd.read_csv("https://bit.ly/3jCROII")
tips
```

```python
# Using the crosstab() function
# To compute a simple cross-tabulation of two (or more) factors.
# Create a 2 factors data frame
party_counts = pd.crosstab(tips["day"], tips["size"])
party_counts
```

```python
# Extract party_count from col 2 to 5
party_counts_edit = party_counts.loc[:, 2:5]
party_counts_edit
```

```python
# Normalize each row to sum to 1
# axis="index" argument indicates that the division should be done along the rows
# some data wrangling of data frame

party_pcts = party_counts_edit.div(party_counts_edit.sum(axis = "columns"), axis = "index")
party_pcts
```

```python
# plot stacked bar
party_pcts.plot.bar(stacked = True);
```

### **Scatter or Point Plots**
examining the relationship between two one-dimensional data series

```python
# Example dataset
# Download the dataset here: https://bit.ly/3FYxCYZ
# Go to this link if you want to know more about this data: http://tinyurl.com/mr3zy799

macro = pd.read_csv("https://bit.ly/3FYxCYZ")
macro
```

```python
# Only select a few variables
# Important for final exam

dat = macro[['cpi', "m1", "tbilrate", "unemp"]]
dat
```

```python
# Log transform the data
# diff() -> calculates the difference between that element and the previous one
# drop.na() because the first element has no previous element to minus
trans_dat = np.log(dat).diff().dropna()
trans_dat
```

```python
# Further explanation
np.log(29.150)-np.log(28.980)
```

```python
# Plot the regression plot
fig = sns.regplot(x = "m1", y = "unemp", data = trans_dat)
fig.set(title = 'Changes in log()m1 vs log(unemp)' );
```

### **Facet Grids and Categorical Data**
